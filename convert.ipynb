```python
import geopy.distance
import pandas as pd
import math as math
import csv
```

```python
data = []

# Read the customer data from the text file
with open('input.txt', 'r') as file:
    # Skip the header line
    next(file)

    # Process each line in the file
    for line in file:
        # Remove leading/trailing whitespace and split the line into attributes
        print(line)
        attributes = line.strip().split()
        print(attributes)
        if len(attributes) >= 3:
        # Convert customer number, X coordinate, and Y coordinate to integers
            xcoord = int(attributes[1])
            ycoord = int(attributes[2])

            # Create a dictionary for the current customer
            entry = {
                'XCOORD.': xcoord,
                'YCOORD.': ycoord
            }

            # Add the dictionary to the data list
            data.append(entry)

# Calculate the distances between customers and store them in a matrix
matrix = []
for i in range(len(data)):
    row = []
    for j in range(len(data)):
        if i == j:
            distance = 0  # Distance from a customer to itself is 0
        else:
            x_diff = data[j]['XCOORD.'] - data[i]['XCOORD.']
            y_diff = data[j]['YCOORD.'] - data[i]['YCOORD.']
            distance = math.sqrt(x_diff ** 2 + y_diff ** 2)
        row.append(distance)
    matrix.append(row)
# print(matrix)
range_column = [i for i in range(401)]
# Create a dataframe from the matrix
df = pd.DataFrame(matrix, index=range_column, columns=range_column)

# Print the dataframe
print(df)
# df.to_csv('distance_matrix.csv', index= False, header=False)

```
